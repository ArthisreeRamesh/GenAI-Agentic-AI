```python
import requests
from bs4 import BeautifulSoup

# Define headers
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3"
}

# Make a GET request to the website with the defined headers
response = requests.get('https://www.flipkart.com/', headers=headers)
html_content = response.text

soup = BeautifulSoup(html_content, 'lxml')

# Removing unnecessary content: script and style tags
for script in soup(["script", "style"]):
    script.extract()

# Get text content
text_content = soup.get_text(strip=True)

# Saving the remaining content
with open("flipkart.txt", "w", encoding="utf-8") as file:
    file.write(text_content)
```

```python
# Let's parse the uploaded flipkart.txt file to extract product categories and their discount percentages.
# We'll look for patterns like "Upto 80% Off" or "Up to 70% Off" and create a table.

import re
import pandas as pd

# Read the file content
file_path = "flipkart.txt"
with open(file_path, "r", encoding="utf-8") as file:
    text = file.read()

# Regex to match items with discounts like "Upto 80% Off" or "Up to 70% Off"
pattern = re.compile(r"([A-Za-z0-9 &]+?)\s*(?:Upto|Up to)\s*(\d+)% Off", re.IGNORECASE)

matches = pattern.findall(text)

# Create a DataFrame
df = pd.DataFrame(matches, columns=["Product/Category", "Discount (%)"])

# Remove duplicates and clean up whitespace
df["Product/Category"] = df["Product/Category"].str.strip()
df = df.drop_duplicates().reset_index(drop=True)

import caas_jupyter_tools
caas_jupyter_tools.display_dataframe_to_user(name="Flipkart Discounts Table", dataframe=df)
```
