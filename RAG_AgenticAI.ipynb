# Define the Agent requirement - LangGraph Workflow

### customer query: text of the customer query
### query_category: {'medical_records', 'appointments', 'billing', 'insurance'}
### query_sentiment: Positive, Negative, Neutral or Distress - (empathetic handling i.e)not just caring or comforting a troubled person but also try to help them)
### escalation_customer_information: human support agent
### oncall_customer_inform -
### Final_response

- CustomerSupportAgentState → this is the state definition (the shared memory that flows through the graph). It’s the “canvas” that all nodes will update.
- QueryCategory / QuerySentiment → these are schemas used inside nodes to validate or parse outputs. They are not nodes themselves.


```python
# Query Category
# Query Sentiment

from typing import TypedDict, Literal #literal - to read str/text values exactly
from pydantic import BaseModel  #pydantic to read diff types of data


class CustomerSupportAgentState(TypedDict):
    customer_query: str
    query_category: str
    query_sentiment: str
    escalation_cust_info: dict
    oncall_cust_info: dict
    final_response: str


class QueryCategory(BaseModel):
  categorized_topic: Literal['medical_records', 'appointments', 'billing', 'insurance']

class QuerySentiment(BaseModel):
  sentiment: Literal['Positive', 'Negative', 'Neutral', 'Distress']
```

```python
# Connect to LLM for our Agent
from langchain_openai import ChatOpenAI
llm = ChatOpenAI(model_name="gpt-4o-mini", temperature=0)
```

- support_state is expected to be a CustomerSupportAgentState object.
- Return type:
-> CustomerSupportAgentState means the function will return a new or updated CustomerSupportAgentState object.



```python
def categorize_inquiry(support_state: CustomerSupportAgentState) -> CustomerSupportAgentState:
  """
  Classify the customer query into {'medical_records', 'appointments', 'billing', 'insurance'}.
  """
  query = support_state["customer_query"]
  ROUTE_CATEGORY_PROMPT = """Act as a customer support agent trying to best categorize the customer query.
                            You are a support agent for a healthcare company focusing a providing healthcare services to customers.
                            These services include:
                            - handle billing queries
                            - handle insurance queries
                            - handle medical records queries
                            - handle appointments queries

                           Please read the customer query below and determine the best category from the following list:
                          'medical_records', 'appointments', 'billing', 'insurance'

                          Remember:
                          - Billing queries will focus more on payment and billing related aspects
                          - Appointments will focus more on booking, rescheduling, cancelling doctor appointments
                          - Records will focus more on updating, sharing, accessing medical records
                          - Insurance will focus more on insurance related queries like claims, updating insurance policy details

                          Return just the category name (from one of the above)

                          Query:
                          {customer_query}
                          """
  prompt = ROUTE_CATEGORY_PROMPT.format(customer_query=query)
  route_category = llm.with_structured_output(QueryCategory).invoke(prompt)
  return {"query_category": route_category.categorized_topic}
```

Create the sentiment analysis

```python
def analyze_inquiery_sentiment(support_state: CustomerSupportAgentState) -> CustomerSupportAgentState:
  """
  Classify the customer query into {'medical_records', 'appointments', 'billing', 'insurance'}.
  """
  query = support_state["customer_query"]
  SENTIMENT_CATEGORY_PROMPT = """Act as a customer support agent trying to best categorize the customer query.
                            You are a support agent for a healthcare company focusing a providing healthcare services to customers.
                            These services include:
                            - handle billing queries
                            - handle insurance queries
                            - handle medical records queries
                            - handle appointments queries

                           Please read the customer sentiment below and determine the best fit from the following list:
                          Positive','Negative','Neutral', or 'Distress'

                          Remember these rules when finding the sentiment:
                          - 'Distress' happens only when the customer is facing a health emergency and might need the on-call emergency doctor
                          - 'Negative' happens only when the customer is not happy with certain products, services offered by the hospital/comapny


                          Return just the category name (from one of the above)

                          Query:
                          {customer_query}
                          """
  prompt = SENTIMENT_CATEGORY_PROMPT.format(customer_query=query)
  sentiment_category = llm.with_structured_output(QuerySentiment).invoke(prompt)
  return {"query_sentiment":sentiment_category.sentiment}
```

Build LangGraph frame by using template method -AGENT IDENTIFY CATEGORY AND RETRIEVE DOCUMENTS REGARDING IT AND Giving response


```python
from langchain_core.prompts import ChatPromptTemplate
from typing import Dict
```

```python
def generate_department_response(support_state: CustomerSupportAgentState) -> CustomerSupportAgentState:
  """
  provide department support response by combing knowledge from the vector store and LLM.
  """
  categorized_topic = support_state["query_category"]
  query = support_state["customer_query"]

  if categorized_topic.lower() =='billing':
    metadata_filter = {"category":"billing"}
    department = "Billing"
  elif categorized_topic.lower() =='insurance':
    metadata_filter = {"category":"insurance"}
    department = "Insurance"
  elif categorized_topic.lower() =='appointments':
    metadata_filter = {"category":"appointments"}
    department = "Appointments"
  elif categorized_topic.lower() =='medical_records':
    metadata_filter = {"category":"medical_records"}
    department = "Medical Records"

  kbase_search.search_kwargs["filter"] = metadata_filter
  relevant_docs = kbase_search.invoke(query)
  relevant_content = "\n\n".join(doc.page_content for doc in relevant_docs)


  prompt = ChatPromptTemplate.from_template(
      """
      Craft a clear and detailed support response for the following customer query about {department}.
      Use the provided knowledge base information to enrich your response.
      In case there is no knowledge base information or you don't know the answer just say:

      Apologies I am not able to help you with that question, please reach out to +91-123432123

      Customer Query:
      {customer_query}

      Relevant Knowledge Base Information:
      {retrieved_content}
      """
  )
  chain = prompt | llm
  reply = chain.invoke({
      "customer_query":query,
      "retrieved_content":relevant_content,
      "department": department
  }).content
  return {"final_response": reply}
```

Step-by-step flow
- Initialize submission flag and state
- Set global flag: form_submitted = False so the loop knows the form hasn’t been submitted yet.
- Prepare result container: result = {} to hold the final values.
- Build the UI
- Header: widgets.HTML shows a title.
- Inputs: Two widgets.Text fields for Name and MobileNo.
- Submit button: widgets.Button(description="Submit").
- Layout: widgets.VBox([...]) stacks header, inputs, and button vertically.
- Render: display(vbox) sends the UI to the notebook frontend.
- Wire the button click handler
- Define on_submit: When called, it:
- Sets form_submitted = True.
- Copies input1.value and input2.value into result['name'] and result['mobileno'].
- Updates button text to “Submitted Successfully”.
- Attach handler: submit_button.on_click(on_submit) so clicking runs the callback.
- Wait for user interaction (non-blocking)
- Enter event-loop context: with ui_events() as poll: integrates with the notebook’s event loop so widget messages are handled while Python code “waits”.
- Poll until submitted: While form_submitted is False, call poll(5) to process events, print a dot for visual feedback, and time.sleep(0.3) to avoid tight looping.
- Return updated agent state
- Output: {'oncall_cust_info': result} so downstream nodes/functions can access the captured user info.
- Why not a simple while loop? A plain blocking loop would prevent the kernel from processing widget events; the click wouldn’t be delivered until the loop ended.
- What ui_events() does: It lets you periodically yield control back to the kernel/frontend so UI events (like button clicks) are handled, making the “pause until submit” pattern work smoothly.
- Polling cadence: poll(5) is the call that processes pending UI events; the short sleep reduces CPU usage while waiting.


- Small value = responsive loop:
A short timeout (like 5 ms) ensures the loop doesn’t block too long. The kernel quickly cycles back, processes events, and prints dots.
- Balance between CPU and responsiveness:
If you set it too low (like poll(0)), the loop might spin too fast and waste CPU.
If you set it too high (like poll(1000)), the UI might feel sluggish because it only checks once per second.
- Combined with poll(5), the loop effectively checks for events every ~0.3 seconds, while still letting the kernel process UI events smoothly.
Summary:
- poll(5) → wait up to 5 ms for widget events.
- Chosen to keep the event loop responsive without wasting CPU.
- Paired with time.sleep(0.3) to create a smooth “heartbeat” while waiting for the user to submit.




```python
def accpet_user_input_escalation(support_state: CustomerSupportAgentState) -> CustomerSupportAgentState:
  """
  provide department support response by combing knowledge from the vector store and LLM
  """
  global form_submitted
  form_submitted = False

  header = widgets.HTML("<h3>Escalation Form - Please enter your details below:<h3>")
  input1 = widgets.Text(description="Name:")
  input2 = widgets.Text(description="MobileNo:")
  input3 = widgets.Text(description="EmailId:")
  result = {}

  def on_submit(submit_button):
    global form_submitted
    form_submitted = True
    result['name'] = input1.value
    result['mobileno'] = input2.value
    result['emailid'] = input3.value
    submit_button.description = "Submitted Successfully"


  submit_button = widgets.Button(description="Submit")
  submit_button.on_click(on_submit)

  # combine all widgets into a vertical layout box
  vbox = widgets.VBox([header, input1, input2, input3, submit_button])
  display(vbox)

  with ui_events() as poll:
    while form_submitted is False:
      poll(5)
      print('.', end='')
      time.sleep(0.3)
  return {
      'escalation_cust_info': result
  }
```

```python
customer_support_graph = StateGraph(CustomerSupportAgentState)

# step 1
customer_support_graph.add_node("categorize_inquiry", categorize_inquiry)
# step 2
customer_support_graph.add_node("analyze_inquiery_sentiment", analyze_inquiery_sentiment)
# step 3
customer_support_graph.add_node("accpet_user_input_escalation", accpet_user_input_escalation)
# step 4
customer_support_graph.add_node("escalate_to_human_agent", escalate_to_human_agent)
# step 5
customer_support_graph.add_node("accpet_user_input_oncall", accpet_user_input_oncall)
# step 6
customer_support_graph.add_node("escalate_to_oncall_team", escalate_to_oncall_team)
# step 7
customer_support_graph.add_node("generate_department_response", generate_department_response)

def determine_route(support_state: CustomerSupportAgentState) -> str:
  """
  determine the next node based on query sentiment and category
  """
  if support_state["query_sentiment"] == "Negative":
    return "accpet_user_input_escalation"
  elif support_state["query_sentiment"] == "Distress":
    return "accpet_user_input_oncall"
  elif support_state["query_category"] in ["billing", "insurance", "appointments", "medical_records"]:
    return "generate_department_response"


customer_support_graph.add_edge("categorize_inquiry", "analyze_inquiery_sentiment")
customer_support_graph.add_conditional_edges(
    "analyze_inquiery_sentiment",
    determine_route,
    ["accpet_user_input_escalation","accpet_user_input_oncall","generate_department_response"]
)

customer_support_graph.add_edge("accpet_user_input_escalation", "escalate_to_human_agent")
customer_support_graph.add_edge("escalate_to_human_agent", END)
customer_support_graph.add_edge("accpet_user_input_oncall", "escalate_to_oncall_team")
customer_support_graph.add_edge("escalate_to_oncall_team", END)
customer_support_graph.add_edge("generate_department_response", END)
customer_support_graph.set_entry_point("categorize_inquiry")


# Compile the graph
'''
from langgraph.
memory = MemorySaver()
form_submitted= False
complied_support_agent = customer_support_graph.compile(memory=memory)
'''
```

Helper Function to RUN the agent

1. agent.stream(...)
- What it is:
This calls the agent’s streaming interface. Instead of returning one final result immediately, the agent yields a sequence of events (usually dictionaries) as it works.
- Arguments passed:
- {"customer_query": prompt} → the user’s input, keyed under the schema field the agent expects.
- {"configurable": {"thread_id": user_session_id}} → metadata telling the agent which conversation/thread this belongs to, so state can be tracked per user.
- stream_mode="values" → tells the agent to yield just the values (plain event dicts) rather than more complex structured objects or debug metadata.
- Analogy:
Think of it like a live broadcast: instead of waiting for the whole show, you get each scene as it happens.


2. final_event = None
- Why we initialize it:
Before looping, we don’t yet know what the agent will produce. Setting final_event = None gives us a safe default.
- During the loop:
As we iterate over events, we keep overwriting final_event with the latest event.
→ After the loop ends, final_event holds the last event (the agent’s final output).
- Analogy:
Like keeping a “last seen message” variable while reading a stream of chat messages.


The condition
if final_event and "final_response" in final_event:
    display(Markdown(final_event["final_response"]))


- Breakdown:
- if final_event → ensures we actually got at least one event (not None).
- "final_response" in final_event → checks that the last event dictionary has a key called "final_response".
- Action:
If both conditions are true, we display the final response nicely formatted in Markdown (so headings, lists, code blocks render properly in the notebook).
- Else branch:
If no final response is found, we print a fallback message: "No final response found."


 Putting it all together
- Start streaming events from the agent.
- Capture each event, keep overwriting final_event.
- After the stream ends, check if the last event has a "final_response".
- If yes → render it in Markdown.
- If not → show a fallback message.
agent.stream() syntax:
agent.stream(
    input_dict,
    config_dict,
    stream_mode="values"
)
Besides "values", you can pass "updates", "messages", or "events" to agent.stream() depending on the level of detail you want from the strea
- values → simplest, just the outputs you care about (final response, tool results).
- updates → good for tracking agent progress step-by-step (retrieval, reasoning, tool calls).
- messages → useful if you want to stream LLM text as it’s generated (like a chatbot typing effect).
- events → best for debugging or logging everything (timestamps, event types, payloads).

🧩 Analogy
Think of stream_mode like camera angles on a live show:
- "values" → just the final scene.
- "updates" → behind-the-scenes progress shots.
- "messages" → close-up of the actor speaking line by line.
- "events" → the director’s full logbook with timestamps and notes.




```python
def call_support_agent(agent, prompt, user_session_id, verbose=False):
  events = agent.stream(
      {"customer_query":prompt},
      {"configurable":{"thread_id": user_session_id}},
      stream_mode="values"
  )
  print("Running Agent. Please wait.....")
  final_event = None
  for event in events:
    if verbose:
      print(event)
    final_event = event


  print("\nFinal Response:")
  if final_event and "final_response" in final_event:
      display(Markdown(final_event["final_response"]))
  else:
      print("No final response found.")
```
